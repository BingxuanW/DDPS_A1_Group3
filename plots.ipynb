## 1. Iterative ML

### 1.1 Different data scale

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# df = pd.read_excel("different_data_scale.xlsx")

def auto_text(rects):
    for rect in rects:
        plt.text(rect.get_x(), rect.get_height(), rect.get_height(), ha='left', va='bottom')

x=np.arange(3)
plt.figure(figsize=(4, 4)) 


y_spark = [30.830, 62.123, 539.934]
y_hadoop = [173.175, 336.841, 1275.666]

bar_width=0.3
x_labels = ['Small', 'Large', 'Huge']

rect1 = plt.bar(x, y_spark, bar_width, color = 'darkseagreen', label = 'Spark')
rect2 = plt.bar(x + bar_width, y_hadoop, bar_width, color = 'lightblue', label = 'Hadoop')

auto_text(rect1)
auto_text(rect2)

plt.legend()  # show legend
plt.xticks(x + bar_width/2, x_labels )
plt.title('Duration of K-means with different Input Data Size')
plt.xlabel('Input data size')
plt.ylabel('Duration(s)')
plt.show()
```

### 1.2 Different Numbers of node

```python
x=np.arange(3)
y_spark = [104.119, 62.123, 58.956]
y_hadoop = [356.419, 336.841, 290.772]

bar_width=0.3
x_labels = ['5', '10', '15']
plt.figure(figsize=(5, 4)) 

rect1 = plt.bar(x, y_spark, bar_width, color = 'darkseagreen', label = 'Spark')
rect2 = plt.bar(x + bar_width, y_hadoop, bar_width, color = 'lightblue', label = 'Hadoop')

auto_text(rect1)
auto_text(rect2)

plt.legend()  # show legend
plt.xticks(x + bar_width/2, x_labels )
plt.title('Duration of K-means with different node numbers under Large Input')
plt.xlabel('Number of nodes')
plt.ylabel('Duration(s)')
plt.show()
```

### 1.3 First iteration vs. subsequent iteration

```python
x=np.arange(2)
plt.figure(figsize=(4, 4)) 

y_spark = [53.348, 1.402]
y_hadoop = [105.451, 46.330]

bar_width=0.3
x_labels = ['First iteration', 'Avg later iteration']

rect1 = plt.bar(x, y_spark, bar_width, color = 'darkseagreen', label = 'Spark')
rect2 = plt.bar(x + bar_width, y_hadoop, bar_width, color = 'lightblue', label = 'Hadoop')

auto_text(rect1)
auto_text(rect2)

plt.legend()  # show legend
plt.xticks(x + bar_width/2, x_labels )
plt.title('First Iteration Duration vs. Later Iteration Average Duration')
plt.xlabel('Iteration Stage')
plt.ylabel('Duration(s)')
plt.show()
```
